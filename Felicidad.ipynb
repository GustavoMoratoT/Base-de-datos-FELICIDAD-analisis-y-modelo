### Modelo

```python
import statsmodels.formula.api as smf
model = smf.ols('Puntaje_de_felicidad ~ PIB_per_capita + Apoyo_social + Esperanza_de_vida_saludable + Libertad_para_tomar_decisiones + Generosidad + Percepcion_de_corrupcion',df).fit()

model.summary()
```

Nuestro modelo presenta un buen ajuste, con un R^2= 0.764  
Mientras que todas las variables implementadas son significativas al 1% menos la Generosidad y la Percepción de Corrupción que lo son al 10% de significancia.
Mencionar que, nuestro modelo no presenta una distribución normal en los residuos, por lo que se detectaran a continuacion...

```python
residuos = model.resid
df["Residuos"] = residuos

df.boxplot("Residuos")
plt.title("Residuos del Modelo")
plt.show()
```

```python
percen25 = df["Residuos"].quantile(0.25)
percen75 = df["Residuos"].quantile(0.75)
iqr = percen75 - percen25
limite_superior = percen75 + 1.5 * iqr
limite_inferior = percen25 - 1.5 * iqr

df2 = df.loc[(df.Residuos<limite_superior) & (df.Residuos>limite_inferior),:]
df2.shape
```

Nuestro DataFrame contaba con 146 filas, por lo que se detectaron 4 valores atipicos en nuestro modelo de regressión lineal multiple y se los eliminó.

```python
model2 = smf.ols('Puntaje_de_felicidad ~ PIB_per_capita + Apoyo_social + Esperanza_de_vida_saludable + Libertad_para_tomar_decisiones + Generosidad + Percepcion_de_corrupcion',df2).fit()

model2.summary()
```

Primeramente, nuestro modelo ha mejorado en su R^2= 0.795.  
Tiene una distribución normal en sus residuos.  
Pero vemos que nuestra variable "Generosidad" no es significativa para el modelo, por lo que se la retirará.

```python
model3 = smf.ols('Puntaje_de_felicidad ~ PIB_per_capita + Apoyo_social + Esperanza_de_vida_saludable + Libertad_para_tomar_decisiones + Percepcion_de_corrupcion',df2).fit()

model3.summary()
```

**Puntaje de Felicidad = 1.9773 + 0.6502 PIB per capita + 1.3842 Apoyo Social + 0.8962 Esperanza de vida saludable + 1.5697 Libertad para tomar decisiones + 0.7016 Percepcion de Corrupcion**

Con nuestro modelo tiene su R^2= 0.793.  
Tiene una distribución normal en sus residuos.  
Todas las viaribles son significativas individualmente y conjutamente.

Tambien destacar dos varibles en nuestro modelo, por un lado, "Apoyo Social", que por cada unidad de apoyo social que tenga un país, incrementa en 1.3842 puntos de felicidad al país. Por otro lado, la variable de "Libertad para tomar decisiones", su beta nos indica que por cada unidad obtenida, el puntaje de felicidad, se incremente en 1.5697 puntos.

Vemos si el modelo presenta multicolinealidad

```python
from patsy import dmatrices
from statsmodels.stats.outliers_influence import variance_inflation_factor

y, X = dmatrices('Puntaje_de_felicidad ~ PIB_per_capita + Apoyo_social + Esperanza_de_vida_saludable + Libertad_para_tomar_decisiones + Percepcion_de_corrupcion', data=df2,return_type='dataframe')
```

```python
vif_df=pd.DataFrame()
vif_df['variable']=X.columns

vif_df['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif_df['TOL'] = [(1/variance_inflation_factor(X.values, i)) for i in range(X.shape[1])]
vif_df['R^2'] = [(1-(1/variance_inflation_factor(X.values, i))) for i in range(X.shape[1])]
```

```python
vif_df
```

Como ninguna de nuestras variables tiene un VIF mayor a 10, se determina que no existe multicolinealidad.

Vemos si el modelo tiene Heterocedasticidad

```python
import statsmodels.stats.api as sm
gq_test = pd.DataFrame(sm.het_goldfeldquandt(model3.resid,model3.model.exog)[:-1],columns=['Value'],index=['F statistic','P-value'])
print(gq_test)
```

```python
from statsmodels.compat import lzip
import statsmodels.stats.api as sms

nombres = ['estadística del multiplicador de Lagrange', 'valor p',
        'valor f', 'valor p f']
test = sms.het_breuschpagan (model3.resid, model3.model.exog)

lzip (nombres, test)
```

Realizando ambas pruebas para detectar la heterocedasticidad, nos indica que no existe homocedasticidad, por que nuestro p-values no es mayor al 0.05.

Realizando las predicciones con los betas obtenidos

```python
caract = ["PIB_per_capita" , "Apoyo_social", "Esperanza_de_vida_saludable" , "Libertad_para_tomar_decisiones" , "Percepcion_de_corrupcion"]
x_valida = df2[caract]

predicciones = model3.predict(x_valida)
df2["predict"] = predicciones

compacion = df2[["Puntaje_de_felicidad", "predict"]]
residuos = model3.resid
compacion["Residuos"] = residuos
```
